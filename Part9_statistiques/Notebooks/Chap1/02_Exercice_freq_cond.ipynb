```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Data tips
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tipsSexSmoker = pd.crosstab(tips['sex'], tips['smoker'], margins=True)
tipsSexSmoker
```

```python
# Fréquences conditionnelles des fumeurs ayant donnés des pourboires
[fSM, fSF]= np.round((tipsSexSmoker['Yes'] / tipsSexSmoker['All'])[:-1], 3)

print(f"fSM = {fSM}, fSF = {fSF}")
```

```python
# Fréquences conditionnelles des non fumeurs ayant donnés des pourboires
[fNoSM, fNoSF]= np.round((tipsSexSmoker['No'] / tipsSexSmoker['All'])[:-1], 3)

print(f"fNoSM = {fNoSM}, fNoSF = {fNoSF}")
```

```python
colors = ['blue', 'orange']
fig, ax = plt.subplots(1,2, figsize=(10, 5))

explode=(0.01, 0.01)

ax[0].pie([fSM, fNoSM], labels=['Smoker', 'No Smoker'], colors=colors, 
        autopct='%1.1f%%', textprops={'color': "black"}, explode=explode)
ax[0].set_title('Male')

ax[1].pie([fSF, fNoSF], labels=['Smoker', 'No Smoker'], colors=colors, 
        autopct='%1.1f%%', textprops={'color': "black"}, explode=explode)
ax[1].set_title('Female')

plt.show();
```

```python
tipsSmokerSex = tipsSexSmoker.T
tipsSmokerSex
```

```python
[fMS, fMNoS]= np.round((tipsSmokerSex['Male'] / tipsSmokerSex['All'])[:-1], 3)

print(f"fMS = {fMS}, fMNoS = {fMNoS}")

[fFS, fFNoS]= np.round((tipsSmokerSex['Female'] / tipsSmokerSex['All'])[:-1], 3)

print(f"fFS = {fFS}, fFNoS = {fFNoS}")
```

```python
# Fréquences conditionnelles des femmes et des hommes ayant donné.es des pourboires
colors = ['blue', 'orange']
fig, ax = plt.subplots(1,2, figsize=(10, 5))

explode=(0.01, 0.01)

# ax tableau numpy
ax[0].pie([fFS, fMS], labels=['Female', 'Male'], colors=colors, 
        autopct='%1.1f%%', textprops={'color': "black"}, explode=explode)

ax[0].set_title('Smoker')

ax[1].pie([fFNoS, fMNoS], labels=['Female', 'Male'], colors=colors, 
        autopct='%1.1f%%', textprops={'color': "black"}, explode=explode)
ax[1].set_title('No Smoker')


plt.show();
```
