# Correction Iris

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
iris.describe()
```

```python
# barplot représentation de l'histogramme fonction de la moyenne
# La bar représente la marge d'erreur dans le calcul de la moyenne, écart type, médiane 
fig = plt.figure(figsize=(10,8))

sns.barplot(data=iris, estimator=np.mean)

plt.xticks(rotation=45);
```

```python
fig = plt.figure(figsize=(10,8))

sns.barplot(data=iris, estimator=np.std)

plt.xticks(rotation=45);
```

```python
fig = plt.figure(figsize=(10,10))
sns.barplot(data=iris, ci=None) 

plt.xticks(rotation=45);
```

```python
# pairplot
sns.pairplot(iris, hue='species', diag_kind='hist', kind='reg');
```

```python
# On peut regarder les corrélation d'une espèce pour ses sepal, petal

iris[ iris.species == 'virginica' ].corr()

# iris.corr()
```

```python
virginica = iris[ iris.species == 'virginica' ]
```

```python
fig = plt.figure(figsize=(10,8))

sns.regplot( x = 'petal_length', y = 'sepal_length', data = virginica, robust=True, ci = None );
```
