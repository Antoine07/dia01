```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
birth = pd.read_csv( './../Data/base_cc_comparateur.CSV', sep=";", low_memory=False)

# CODGEO P18_POP NAISD20
birth.head()
```

```python
birth['CODGEO'] = birth.CODGEO.astype('str')
```

```python
birth['CODGEO'] = birth['CODGEO'].str[:2]

birth.head()
```

```python
birth = birth[ birth["CODGEO"].isin(["77","78","91","92","93","94","95"]) ][["CODGEO", "P18_POP", "NAISD20"]]
birth['ratio'] = np.round( (birth['NAISD20'] / birth['P18_POP']) * 1_000, 2)

birth.head()
```

```python
fig_dims = (10, 15)
fig, ax = plt.subplots(figsize=fig_dims)

sns.boxplot(x="CODGEO", y="ratio", data=birth, showfliers=False);
```
